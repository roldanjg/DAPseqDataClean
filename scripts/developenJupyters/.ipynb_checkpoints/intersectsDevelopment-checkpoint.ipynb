```python
import pandas as pd

boxes = ['G_PBE_merged_intersect.csv','PBE_merged_intersect.csv', 'Gbox_merged_intersect.csv']

reponeNames = pd.read_csv('rep1.txt', names=['inde','name','reads','percentage'])
reptwoNames = pd.read_csv('rep2.txt', names=['inde','name','reads','percentage'])

for case in  boxes:
    repOnedf = pd.read_csv(f'rep1/{case}').fillna(0.0)
    reptwodf = pd.read_csv(f'rep2/{case}').fillna(0.0)
    print(case)
    for tf in ['PIF4sWT','PIF4sloopMYC2','PIF5sWT','PIF5sloopMYC2','MYC2sWT','MYC2sloopPIF4','MYC2sloopPIF5']:
        
        readstwo = reptwoNames[reptwoNames['name'] == tf]['reads'].values[0]
        percentwo = reptwoNames[reptwoNames['name'] == tf]['percentage'].values[0]
        normalizationPerReadTwo = readstwo * percentwo
        
        readsone = reponeNames[reponeNames['name'] == tf]['reads'].values[0]
        percenone = reponeNames[reponeNames['name'] == tf]['percentage'].values[0]
        normalizationPerReadTwo = readsone * percenone
        
#         ########
        if tf == 'MYC2sWT':
            # in case we are using MYC2sWT, divide by the Rep2 input, becouse both are from this replicate
            repOnedf[f'rep1_{tf}'] =  (repOnedf[f'{tf}']/repOnedf[f'{tf}'].sum()) /  (reptwodf['InputInput']/reptwodf['InputInput'].sum())
            reptwodf[f'rep2_{tf}'] =  (reptwodf[f'{tf}']/reptwodf[f'{tf}'].sum()) /  (reptwodf['InputInput']/reptwodf['InputInput'].sum())
        else:
            repOnedf[f'rep1_{tf}'] =  (repOnedf[f'{tf}']/repOnedf[f'{tf}'].sum()) /  (repOnedf['InputInput']/repOnedf['InputInput'].sum())
            reptwodf[f'rep2_{tf}'] =  (reptwodf[f'{tf}']/reptwodf[f'{tf}'].sum()) /  (reptwodf['InputInput']/reptwodf['InputInput'].sum())
        repOnedf.drop(columns=[f'{tf}'], inplace=True)
        reptwodf.drop(columns=[f'{tf}'], inplace=True)
        
#         print(repOnedf[f'{tf}'], repOnedf['InputInput'], reptwodf[f'{tf}'], reptwodf['InputInput'])
    finaldf = repOnedf.merge(reptwodf, on=['chr', 'start', 'end', 'boxname'], how='outer')
    for tf in ['PIF4sWT','PIF4sloopMYC2','PIF5sWT','PIF5sloopMYC2','MYC2sWT','MYC2sloopPIF4','MYC2sloopPIF5']:
        finaldf[tf] = (finaldf[f'rep1_{tf}']+finaldf[f'rep2_{tf}'])/2
        finaldf.drop(columns=[f'rep1_{tf}'], inplace=True)
        finaldf.drop(columns=[f'rep2_{tf}'], inplace=True)
    finaldf.to_csv(f'{case}_MEAN_REPLICATES_TOTALBOXES',sep='\t', index=False)
# normalization using total count
#         totalColumn = heatdf[column].sum()
#         lencolumn = len(heatdf[column])
#         meanCollumn = totalColumn/lencolumn
#         heatdf[column] = heatdf[column].map(lambda boxSum: math.log2(boxSum/meanCollumn))
        
# normalization using reads per million

         
        
#     print(repOnedf, reptwodf)
```
